```python
# Seleccionar la fila 391 (visualizamos el cambio que hubo del codigo de arriba)
row_391 = df_grocery.iloc[389]  # Se resta 1 porque el índice comienza en 0

# Mostrar la fila seleccionada
print(row_391)
```

```python
# Hay muchas filas con el mismo problema por lo que debemos aplicar una solucion que le afecte a todas las filas (habra que volver a ejecutar el codigo que cambia los valores de str a float)
import re # Librería para trabajar con expresiones regulares = Patrones de búsqueda para encontrar coincidencias en cadenas de texto

# Definir una función para limpiar la columna Price
def clean_price(price):
    # Verificar si el valor es una cadena de texto
    if isinstance(price, str): # Formato de isinstance --> isinstance(objeto, tipo) sirve para verificar si un valor es de cierto tipo de dato
        # Buscar el patrón "$X.XXthrough-$Y.YY"
        pattern = r'\$[\d.]+through-\$[\d.]+'
        match = re.search(pattern, price) # search(pattern, string): Esta función busca en la cadena de texto (string) la primera ocurrencia del patrón especificado (pattern). Si se encuentra una coincidencia, devuelve un objeto de tipo "Match", que contiene información sobre la posición de la coincidencia y el texto coincidente. Si no se encuentra ninguna coincidencia, devuelve None.
        if match:
            # Si se encuentra el patrón, devolver solo el primer valor ($X.XX)
            return match.group(0).split('through')[0] # match.group(0): Esta parte del código obtiene la cadena de texto que coincide con el patrón encontrado. El método group(0) devuelve la cadena completa que coincide con el patrón.
            # .split('through')[0]: Luego, esta parte divide la cadena de texto en dos partes utilizando el texto "through" como punto de división y devuelve la primera parte. Esto es porque el patrón tiene la forma "$X.XXthrough-$Y.YY", donde "through" separa dos valores de precio. Al dividir la cadena en estas partes y seleccionar solo la primera parte (antes de "through"), estamos obteniendo solo el primer valor de precio ($X.XX).
        else:
            # Si no se encuentra el patrón, devolver el valor original
            return price
    else:
        # Si el valor no es una cadena de texto, devolver el valor original
        return price

# Aplicar la función a la columna Price
df_grocery['Price'] = df_grocery['Price'].apply(clean_price)
```

### Explicación del patrón "$X.XXthrough-$Y.YY"

El patrón "r'\$[\d.]+through-\$[\d.]+" se utiliza para encontrar cadenas de texto que siguen un formato específico en el cual hay dos números decimales separados por la palabra "through", ambos precedidos por el signo de dólar "$".

Aquí está desglosado el significado de cada parte del patrón:

**r:** Indica que la cadena de texto es una "raw string" en Python, lo que significa que los caracteres de escape no tienen ningún significado especial, lo que simplifica la escritura de expresiones regulares.

**\$:** Busca el símbolo del dólar "$".
El símbolo de dolar se escapa con una barra invertida para indicar que se debe buscar literalmente el carácter

**[\d.]+:** Busca uno o más dígitos numéricos y/o el punto decimal ".". El corchete "[" y "]" define un conjunto de caracteres, mientras que "\d" coincide con cualquier dígito decimal y "." coincide con el punto decimal. El signo "+" indica que el conjunto de caracteres puede aparecer una o más veces.

**through-:** Busca la palabra "through-" literalmente. Esto indica que la cadena de texto debe contener esta palabra específica después del primer número y antes del segundo número.

**\$[\d.]+:** Igual que la primera parte, busca otro número decimal precedido por el símbolo del dólar "$".

**+:** Indica que la parte anterior de la expresión regular (que incluye el número y el "through") puede aparecer una o más veces.

En resumen, este patrón busca cadenas de texto que tengan la forma "$X.XXthrough-$Y.YY", donde "X.XX" y "Y.YY" son números decimales separados por "through-".

```python
import pandas as pd
import re

# Convertir la columna 'Rating' a tipo cadena
df_grocery_clean['Rating'] = df_grocery_clean['Rating'].astype(str)

# Definir una función para extraer la calificación y la cantidad de reviews
def extract_rating(review_text):
    if review_text == 'No Reviews':
        return None, None
    else:
        # Usar expresiones regulares para extraer la calificación y la cantidad de reviews
        pattern = r'Rated ([\d.]+) out of 5 stars based on (\d+) reviews\.'
        match = re.match(pattern, review_text)
        if match:
            rating = float(match.group(1))
            reviews = int(match.group(2))
            return rating, reviews
        else:
            return None, None

# Aplicar la función a la columna 'Rating' de df_grocery_clean y crear columnas nuevas
df_grocery_clean['Rating_Value'], df_grocery_clean['Reviews'] = zip(*df_grocery_clean['Rating'].map(extract_rating))
```

```python
# Calcular el promedio de Rating_Value por cada subcategoría
promedio_rating_por_subcategoria = df_grocery_clean.groupby('Sub Category')['Rating_Value'].mean().sort_values(ascending=False) # Lo ordenamos de mayor a menor

# Mostrar los resultados
print(promedio_rating_por_subcategoria)
```

```python
# Crear la gráfica de barras
plt.figure(figsize=(10, 6))
promedio_rating_por_subcategoria.plot(kind='bar', color='skyblue')
plt.title('Promedio de Rating_Value por Subcategoría')
plt.xlabel('Subcategoría')
plt.ylabel('Promedio de Rating_Value')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Calcular la correlación entre Rating_Value y Reviews ver si existe alguna relación entre la calificación promedio y la cantidad de reseñas recibidas para cada producto.
correlation = df_grocery_clean['Rating_Value'].corr(df_grocery_clean['Reviews'])

print("Correlación entre Rating_Value y Reviews:", correlation)
```



*   Si la correlación es cercana a 1, significa que hay una correlación positiva fuerte entre las dos variables, lo que indica que cuando una variable aumenta, la otra también tiende a aumentar.
*   Si la correlación es cercana a -1, significa que hay una correlación negativa fuerte entre las dos variables, lo que indica que cuando una variable aumenta, la otra tiende a disminuir.
*   Si la correlación es cercana a 0, significa que no hay una correlación lineal fuerte entre las dos variables.
